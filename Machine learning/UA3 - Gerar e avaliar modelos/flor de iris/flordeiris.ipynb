<p>O conjunto de dados Iris contém informações de 150 amostras de plantas Iris divididas em três espécies: Setosa, Versicolor e Virgínica, com 50 amostras por espécie. Cada amosra é caracterizada por 4 atributos relacionados ao comprimento, largura e altura das sépalas e pétalas. Esses dados são frequentement utilizados em tarefas de classificação e clusterização tendo um exemplo clássico na Análise de Dados e Aprendizado de Máquina.</p>

<h3>Machine Learning e o Iris Dataset</h3>
<p>Vamos aplicar uma solução de <b>APRENDIZADO SUPERVISIONADO</b> para descobrir que espécie de Iris é.</p>
<p>Para a classificação, vamos utilizar os seguintes atributos:</p>
<ol>
    <li>Altura e largura das pétalas</li>
    <li>Altura e largura das sépalas</li>
</ol>

<h3>Passo a passo:</h3>
<h4>Importando as libs</h4>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

# mostrando os dados no Jupyter
%matplotlib inline
```

<h4>Carregando CSV em um Dataframe e apresentando os dados</h4>

```python
df = pd.read_csv('./data/iris.csv')
df
```

<h4>Plotando os dados</h4>

```python
sb.pairplot(df, hue="target")
```

<p>O que o gráfico acima representa? Mostra como está a dispersão combinando características com características. Com essa classificação, claramente será fácil do nosso software distinguir qual espécie será quando aplicarmos uma nova.</p>

<h4>Selecionando as features para classificação Numpy</h4>

```python
X = np.array(df.drop('target', axis=1))
Y = np.array(df.target)
```

<h3>Algorítimo de KNN</h3>
<p>O algoritmo KNN (K-Nearest Neighbors) é um método de classificação e regressão em aprendizado de máquina. Ele determina a classe ou valor de uma amostra com base na maioria dos k vizinhos mais próximos, medindo a distância entre eles. É um algoritmo simples de implementar e eficaz para problemas de baixa dimensionalidade, mas pode ser sensível à escolha de k e à escala dos dados.</p>

<img src="./statics/knn.jpg">

<h4>Importando o KNN</h4>

```python
from sklearn.neighbors import KNeighborsClassifier
```

<h4>Criando um classificador</h4>

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

<h4>Gerando um treinamento</h4>

```python
knn.fit(X,Y)
```

<h4>Gerando predição para novas classificações</h4>

```python
knn.predict([[6.5,6.5,4.7,1.3]])
knn.predict([[4.2,3.1,2.1,0.9]])
```

<h4>Salvando o resultado</h4>

```python
import joblib

joblib.dump(knn, './res/iris_modelo.pkl')
```

<h4>Carregando o modelo do resultado e gerando uma nova predição</h4>

```python
import joblib

knn2 = joblib.load('./res/iris_modelo.pkl')
knn2.classes_
knn2.predict([[4.2,3.1,2.1,5.9]])
```
