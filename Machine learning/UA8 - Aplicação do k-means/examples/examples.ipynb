```python
# Exemplo 6
# Carrega bibliotecas e módulos necessários
from sklearn import datasets as ds
from sklearn.cluster import KMeans
import pandas as pd  # Corrigi a importação da biblioteca pandas
import seaborn as sb
import matplotlib.pyplot as pp

# Distribuição aleatória: 100 objetos, 4 grupos, 2 características, desvio padrão de 1.3
X, y = ds.make_blobs(n_samples=100, n_features=2, centers=4, cluster_std=1.3)

# Inicializa e treina um algoritmo de K-Means
km = KMeans(n_clusters=4, random_state=50)
km.fit(X)
prevs = km.predict(X)

# DataFrames para o conjunto de dados original com seu atributo alvo e outro quadro
# com o atributo alvo previsto a partir do K-Means
xDF = pd.DataFrame(data=X, columns=['Attr1', 'Attr2'])
xDF['Target'] = y

pDF = pd.DataFrame(data=X, columns=['Attr1', 'Attr2'])  # Corrigi a criação do DataFrame pDF
pDF['Target'] = prevs

# Divisão da figura em duas colunas para plotar os dois conjuntos
fig, axes = pp.subplots(1, 2, figsize=(12, 4))
sb.scatterplot(data=xDF, x='Attr1', y='Attr2', hue='Target', palette='rainbow', ax=axes[0])
sb.scatterplot(data=pDF, x='Attr1', y='Attr2', hue='Target', palette='rainbow', ax=axes[1])
```

```python
# Exemplo 7
# Importação das bibliotecas e seleção dos módulos a serem utilizados
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import pandas as pd  # Corrigi a importação da biblioteca pandas
import seaborn as sb
import matplotlib.pyplot as pp
import numpy as np

# Importação e separação dos dados
iDS = load_iris()
iData, iTarget = iDS.data, iDS.target  # Corrigi a atribuição dos dados

# Formatação dos dados em quadro da biblioteca pandas para facilitar a visualização
iFeature_Names = iDS.feature_names
iDF = pd.DataFrame(data=iData, columns=iFeature_Names)
iDF_Target = iDF.copy()
iDF_Target['species'] = iDS.target_names[iTarget]  # Corrigi a atribuição do nome das espécies

# Visualizar a sobreposição das espécies entre todas as combinações de pares de características
sb.pairplot(data=iDF_Target, hue='species')

# Exibição do gráfico
pp.show()
```

```python
# Seleção das duas características
iData_SL_PW = iData[:, [0, 3]]
iDF_SL_PW = pd.DataFrame(data=iData_SL_PW, columns=[iFeature_Names[0], iFeature_Names[3]])
iDF_SL_PW_T = iDF_SL_PW.copy()
iDF_SL_PW_T['species'] = iDF_Target['species']

# Distribuição após seleção dos atributos sem rótulos definidos
prevFeatureNames = iDF_SL_PW_T.columns
sb.scatterplot(data=iDF_SL_PW_T, x=prevFeatureNames[0], y=prevFeatureNames[1])

# Exibição do gráfico
pp.show()
```

O algoritmo de k-means é configurado para buscar a mesma quantidade de clusters que a quantidade de espécies esperada, a previsão de agrupamento é treinada
e feita segundo os clusters encontrados para o mesmo conjunto de dados utilizados
no treinamento. O vetor resultante é atribuído ao novo DataFrame, agora com
os dados das duas características de entrada, mas como atributo alvo os valores
previstos pelo algoritmo. A seguir, é apresentado o código desse procedimento.

```python
# Exemplo 9
# Quantidade de clusters será a mesma da quantidade de espécies do conjunto de dados
k = len(iDS.target_names)

# Configuração do algoritmo e treinamento
alg = KMeans(n_clusters=k, random_state=50)
alg.fit(X=iData_SL_PW)

# Construção de novo quadro considerando os agrupamentos previstos pelo K-Means
prevDataFrame = iDF_SL_PW.copy()
prevDataFrame['predicted_specie'] = alg.predict(iData_SL_PW)

# Visualização dos resultados usando um gráfico de dispersão
sb.scatterplot(data=prevDataFrame, x=prevFeatureNames[0], y=prevFeatureNames[1], hue='predicted_specie', palette='viridis')
pp.show()

# Exibição do DataFrame com os agrupamentos previstos
prevDataFrame
```

Uma matriz de confusão é uma tabela que é frequentemente utilizada para avaliar o desempenho de um modelo de classificação em problemas de aprendizado supervisionado. Ela compara as previsões feitas pelo modelo com os valores reais conhecidos.

A matriz de confusão é particularmente útil quando o problema de classificação envolve mais de duas classes, permitindo a análise detalhada de como o modelo está performando em cada classe.

A estrutura básica de uma matriz de confusão para um problema de classificação com duas classes (positivo e negativo) é a seguinte:


TP (True Positive): Número de casos em que o modelo previu corretamente a classe positiva.
FP (False Positive): Número de casos em que o modelo previu incorretamente a classe positiva (quando a classe real é negativa).
FN (False Negative): Número de casos em que o modelo previu incorretamente a classe negativa (quando a classe real é positiva).
TN (True Negative): Número de casos em que o modelo previu corretamente a classe negativa.
A partir dessa matriz, várias métricas de desempenho podem ser calculadas, incluindo:

Acurácia (Accuracy): (TP + TN) / (TP + FP + FN + TN)
Precisão (Precision): TP / (TP + FP)
Recall (Sensibilidade ou Taxa de Verdadeiros Positivos): TP / (TP + FN)
F1-Score: Uma métrica que combina precisão e recall para fornecer um equilíbrio entre ambas.
Essas métricas ajudam a fornecer uma visão mais completa do desempenho do modelo em diferentes aspectos, considerando tanto os erros quanto os acertos em diferentes categorias.
